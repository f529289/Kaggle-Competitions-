```python
import pandas as pd
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from textblob import TextBlob

# Download NLTK data (if not already downloaded)
nltk.download('punkt')
nltk.download('stopwords')

# Function to perform keyword extraction
def extract_keywords(text):
    words = word_tokenize(text.lower())  # Convert to lowercase
    stop_words = set(stopwords.words('english'))
    filtered_words = [word for word in words if word.isalnum() and word not in stop_words]
    word_freq = nltk.FreqDist(filtered_words)
    keywords = [word for word, freq in word_freq.most_common(5)]  # Adjust the number as needed
    return keywords

# Function to perform sentiment analysis
def analyze_sentiment(text):
    analysis = TextBlob(text)
    sentiment = analysis.sentiment.polarity  # Polarity ranges from -1 (negative) to 1 (positive)
    return sentiment

# Apply keyword extraction and sentiment analysis to the 'desc' column
df['keywords'] = df['desc'].apply(extract_keywords)
df['sentiment'] = df['desc'].apply(analyze_sentiment)

# Add the sentiment scores to a new column 'sentiment_score'
df['sentiment_score'] = df['sentiment']

# Print the DataFrame with keywords and the new 'sentiment_score' column
print(df[['name', 'desc', 'keywords', 'sentiment_score']])
```


Choosing the "best" combination of five types for Pok√©mon team coverage depends on various factors, including the current metagame, the specific Pok√©mon you have available, and your preferred battling style. However, for a balanced and versatile team, you can aim for a combination of types that collectively cover a wide range of strengths and weaknesses. Here's one such combination:

**Water Type:** Water types provide excellent coverage, resistances, and versatility. They are weak to Electric and Grass types but resist Water, Fire, Ice, and Steel types.

**Grass Type:** Grass types are effective against Water and Ground types, providing strong coverage against those common threats. They are weak to Fire, Ice, Poison, Flying, and Bug types.

**Fire Type**: Fire types are strong against Grass, Bug, Ice, and Steel types. They have weaknesses to Water, Rock, and Ground types.

**Electric Type:** Electric types are powerful offensively and have only one weakness (Ground). They resist Electric and Flying types.

**Fairy Type:** Fairy types can deal with Dragon types effectively and have good resistance to Bug, Dark, and Fighting types. They are weak to Steel and Poison types.

```python
# Assuming you have a DataFrame named 'df' with the specified columns.

# List of types to search for (in lowercase).
types_to_search = ['water', 'grass', 'fire', 'electric', 'fairy']

# Create an empty dictionary to store the top five Pok√©mon for each type.
top_five_pokemon_by_type = {}

# Iterate through each type and find the top five Pok√©mon with the highest total base stat.
for type_to_search in types_to_search:
    # Filter the DataFrame for Pok√©mon of the current type in either type1 or type2.
    filtered_pokemon = df[(df['type1'].str.lower() == type_to_search) | (df['type2'].str.lower() == type_to_search)]

    # Check if the filtered DataFrame is not empty.
    if not filtered_pokemon.empty:
        # Sort the DataFrame by total base stat in descending order and take the top five.
        top_five_pokemon = filtered_pokemon.sort_values(by='total', ascending=False).head(5)

        # Add the top five Pok√©mon for the current type to the dictionary.
        top_five_pokemon_by_type[type_to_search] = top_five_pokemon
    else:
        # Handle the case when there are no Pok√©mon of the current type.
        top_five_pokemon_by_type[type_to_search] = None

# Print the results.
for type_name, top_five_pokemon in top_five_pokemon_by_type.items():
    if top_five_pokemon is not None:
        print(f"Top five {type_name} Pok√©mon:")
        for index, row in top_five_pokemon.iterrows():
            pokemon_types = [row['type1'], row['type2']] if not pd.isna(row['type2']) else [row['type1']]
            print(f"   {row['name']} (Types: {', '.join(pokemon_types)} | Total Stats: {row['total']})")
    else:
        print(f"No {type_name} Pok√©mon found in the dataset.")
```

```python
import pandas as pd
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from textblob import TextBlob
import matplotlib.pyplot as plt
import seaborn as sns

# Assuming you have a DataFrame named 'df' with the 'desc', 'type1', and 'type2' columns.
# Replace 'df' with the actual name of your DataFrame.

# Download NLTK data (if not already downloaded)
nltk.download('punkt')
nltk.download('stopwords')

# Function to perform sentiment analysis
def analyze_sentiment(text):
    analysis = TextBlob(text)
    sentiment = analysis.sentiment.polarity  # Polarity ranges from -1 (negative) to 1 (positive)
    return sentiment

# Apply sentiment analysis to the 'desc' column
df['sentiment'] = df['desc'].apply(analyze_sentiment)

# Combine 'type1' and 'type2' columns to create a new column 'combined_type'
df['combined_type'] = df['type1'] + ' / ' + df['type2']

# Calculate the top 10 mean sentiment scores by 'combined_type'
top_10_combined_types = df.groupby('combined_type')['sentiment'].mean().sort_values(ascending=False).head(10).index

# Filter the DataFrame to include only the top 10 combined types
df_top_10 = df[df['combined_type'].isin(top_10_combined_types)]

# Create a bar graph to show the mean sentiment scores for the top 10 combined types
plt.figure(figsize=(12, 6))
sns.barplot(x='combined_type', y='sentiment', data=df_top_10, order=top_10_combined_types)
plt.title('Top 10 Mean Sentiment Scores by Combined Type (type1 / type2)')
plt.xlabel('Combined Type (type1 / type2)')
plt.ylabel('Mean Sentiment Score')
plt.xticks(rotation=45, ha='right')  # Rotate x-axis labels for readability

# Show the plot
plt.tight_layout()
plt.show()
```

```python
import pandas as pd

# Assuming you have a DataFrame named 'df' with the 'name', 'sentiment', and 'total' columns.
# Replace 'df' with the actual name of your DataFrame.

def find_top_pokemon(df, num_top=5):
    # Sort the DataFrame by sentiment (in descending order) and total stats (in descending order)
    sorted_df = df.sort_values(by=['sentiment', 'total'], ascending=[False, False])

    # Get the information of the top Pok√©mon
    top_pokemon_info = sorted_df.iloc[:num_top]

    # Extract the relevant information for the top Pok√©mon
    top_pokemon_names = top_pokemon_info['name'].tolist()
    top_pokemon_sentiments = top_pokemon_info['sentiment'].tolist()
    top_pokemon_totals = top_pokemon_info['total'].tolist()

    return top_pokemon_names, top_pokemon_sentiments, top_pokemon_totals

# Example usage to show the top 5 Pok√©mon
top_pokemon_names, top_pokemon_sentiments, top_pokemon_totals = find_top_pokemon(df, num_top=5)

print("Top 5 Pok√©mon:")
for i in range(len(top_pokemon_names)):
    print(f"Name: {top_pokemon_names[i]}")
    print(f"Sentiment Score: {top_pokemon_sentiments[i]}")
    print(f"Total Stat: {top_pokemon_totals[i]}")
    print()
```
