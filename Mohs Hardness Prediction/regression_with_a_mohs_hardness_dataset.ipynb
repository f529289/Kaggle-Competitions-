```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import median_absolute_error
from sklearn.preprocessing import StandardScaler
import pandas as pd

# Assuming 'df' is your DataFrame
# Assuming 'Hardness' is the target variable

# Extract features and target variable
X = df[['allelectrons_Total', 'density_Total', 'allelectrons_Average',
        'val_e_Average', 'atomicweight_Average', 'ionenergy_Average',
        'el_neg_chi_Average', 'R_vdw_element_Average', 'R_cov_element_Average',
        'zaratio_Average', 'density_Average']]
y = df['Hardness']

# Scale the features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42)

# Initialize the linear regression model
model = LinearRegression()

# Fit the model on the training data
model.fit(X_train, y_train)

# Make predictions on the test set
y_pred = model.predict(X_test)

# Evaluate the model using median absolute error
med_abs_error = median_absolute_error(y_test, y_pred)
print(f'Median Absolute Error: {med_abs_error}')

```

#Multicollineraity

```python
import pandas as pd
from statsmodels.stats.outliers_influence import variance_inflation_factor

# Assuming 'df' is your DataFrame
# Assuming 'Hardness' is the target variable
# Extract features
X = df[['allelectrons_Total', 'density_Total', 'allelectrons_Average',
        'val_e_Average', 'atomicweight_Average', 'ionenergy_Average',
        'el_neg_chi_Average', 'R_vdw_element_Average', 'R_cov_element_Average',
        'zaratio_Average', 'density_Average']]

# Calculate VIF for each feature
vif_data = pd.DataFrame()
vif_data["Feature"] = X.columns
vif_data["VIF"] = [variance_inflation_factor(X.values, i) for i in range(X.shape[1])]

# Display the VIF data
print(vif_data)
```

```python
import numpy as np

# Set the correlation threshold
correlation_threshold = 0.8

# Create a mask for features to keep
corr_matrix = X.corr().abs()
upper = corr_matrix.where(np.triu(np.ones(corr_matrix.shape), k=1).astype(bool))
to_drop = [column for column in upper.columns if any(upper[column] > correlation_threshold)]

# Drop highly correlated features
X_filtered = X.drop(columns=to_drop)
```
