```python
# Sample data: List of countries, UFO sightings, and populations (you can replace this with your data)
countries = ["USA", "Canada", "Germany", "UK", "Australia"]
sightings = [70293, 3266, 112, 2050, 593]
populations = [331002651, 38005238, 83783942, 67886011, 25499884]

# Calculate sightings per capita for each country #ChatGPT
sightings_per_capita = [sightings[i] / populations[i] for i in range(len(countries))]

# Calculate the average sightings per capita across all countries
average_sightings_per_capita = sum(sightings_per_capita) / len(countries)

# Normalize the sightings based on the average sightings per capita #ChatGPT helped
normalized_sightings = [sightings[i] * (average_sightings_per_capita / sightings_per_capita[i]) for i in range(len(countries))]

# Print the results
print("Original Sightings:", sightings)
print("Normalized Sightings:", [int(round(x)) for x in normalized_sightings])
```

```python
import matplotlib.pyplot as plt

# Create a line plot for the original sightings
plt.plot(countries, sightings, label='Original Sightings', marker='o', linestyle='-', color='b')

# Create a line plot for the normalized sightings
plt.plot(countries, normalized_sightings, label='Normalized Sightings', marker='s', linestyle='--', color='r')

# Add labels and a legend
plt.xlabel('Countries')
plt.ylabel('Number of Sightings')
plt.title('UFO Sightings vs. Normalized UFO Sightings by Country')
plt.legend()

# Rotate the x-axis labels for better readability (optional)
plt.xticks(rotation=45)

# Show the plot
plt.grid(True)
plt.tight_layout()
plt.show()
```
